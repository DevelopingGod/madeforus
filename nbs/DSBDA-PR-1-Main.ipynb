## Operations needed:-
### 1) Make, fuel-type, aspiration, body-style, drive-wheels, engine-location, engine-type, fuel-system are categorical values that need to be converted to quantitative.
### 2) num-of-doors and num-of-cylinders have values present but in text so need to handle that

```python
# Now we have to deal with:- 
# Make, fuel-type, aspiration, body-style, drive-wheels, engine-location, engine-type, 
# fuel-system are categorical values that need to be converted to quantitative.
```
