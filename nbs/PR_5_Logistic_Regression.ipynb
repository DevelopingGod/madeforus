## Link to Dataset: https://www.kaggle.com/datasets/akram24/social-network-ads/data

### **About:** Dataset on Social media ads describing users, whether users have purchased a product by clicking on the advertisements shown to them.

```python
# Importing the required libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Beyond the practical
from sklearn.naive_bayes import GaussianNB
```

```python
# Reading the data
df = pd.read_csv("Social_Network_Ads.csv")
df.head(8)
```

```python
# checking basic statistics
print(df.shape)
df.info()
```

```python
df.describe()
```

```python
# Drop the unwanted column User ID as it is irrelevent
df.drop(['User ID'],axis=1,inplace=True)
```

```python
df.columns
```

```python
# To check the class imbalance
df.Purchased.value_counts()
```

```python
# Gender is object column so use imputation
label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])  # Male=1, Female=0
```

```python
x=df[['Gender','Age','EstimatedSalary']]
x.head()
```

```python
y=df['Purchased']
y.head()
```

```python
xtrain,xtest, ytrain, ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
# Fitting data into model
model=LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
comparer = pd.DataFrame({'Actual': ytest, 'Predicted': ypred})
comparer.head(10)
```

```python
# Evaluating the performance of the model
accuracy=accuracy_score(ytest, ypred)
cm=confusion_matrix(ytest,ypred)
TN, FP, FN, TP = cm.ravel()
report=classification_report(ytest, ypred)
print("Accuracy:", accuracy)
print("Confusion Matrix:")
print(cm)
print("True Negatives (TN):", TN)
print("False Positives (FP):", FP)
print("False Negatives (FN):", FN)
print("True Positives (TP):", TP)
```
